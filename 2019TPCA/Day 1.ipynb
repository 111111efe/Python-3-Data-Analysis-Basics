```python
x = np.linspace(0, 1.5, 100)
y = np.sin(x)
```

```python
plt.plot(x,y)
```

```python
Y = np.sin(x) + 0.4*np.random.randn(100)
```

```python
plt.scatter(x, Y)
plt.plot(x, y, 'r')
```

```python
def myplot(k):
    Y = np.sin(x) + k*np.random.randn(100)
    plt.ylim(-0.5,2)
    plt.scatter(x, Y)
    plt.plot(x, y, 'r')
```

```python
interact(myplot, k=(0.1, 0.5))
```
